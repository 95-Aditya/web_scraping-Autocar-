```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
a=requests.get("https://www.autocarindia.com/cars/cars-between-5-to-10-lakhs")
a
```

```python
soup=BeautifulSoup(a.content,'html.parser')
soup
```

```python
names=soup.find_all('h3',class_="heading-h4")
name=[]
for i in names:
    n=i.get_text()
    name.append(n)
name
```

```python
prices=soup.find_all('div',class_="ac-price1 ac-mr-right")
price=[]
for i in prices:
    p=i.get_text()
    price.append(p)
price 
```

```python
engines=soup.find_all('p',class_='')
engine=[]
for i in engines:
    e=i.get_text()
    engine.append(e)
engine[0:37]
```

```python
unwanted=[1,2,3,5,6,7,9,10,11,13,14,15,17,18,19,21,22,23,25,26,27,29,30,31,33,34,35]
for ele in sorted(unwanted,reverse=True):
    del engine[ele]
```

```python
engine_cc=engine[0:10]
engine_cc
```

```python
ratings=soup.find_all('span',class_="")
rating=[]
for i in ratings[2:12]:
    r=i.get_text()
    rating.append(r)
rating    
```

```python
images=soup.find_all('img',class_="img-fluid listing-img-car")
image=[]
for i in images:
    im=i['src']
    image.append(im)
image       
```

```python
df=pd.DataFrame()
df
```

```python
df["Name"]=name
df["Price"]=price
df["Engine(cc)"]=engine_cc
df["Rating"]=rating
df["Image"]=image
df
```

```python
df.to_csv("Budget_Cars.csv")
```
